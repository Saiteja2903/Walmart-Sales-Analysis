
# Sales Contribution by Category

```
category_sales = df.groupby('Category')['Sales'].sum().sort_values(ascending=False)
print(category_sales)
plt.figure(figsize=(8, 6))
category_sales.plot( kind='bar', color='skyblue')
plt.title('Sales Contribution by Category')
plt.ylabel('Total Sales')
plt.xlabel('Category')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Discount', y='Profit', hue='Category',alpha=0.7)
plt.title('Profitability vs. Discounts')
plt.xlabel('Discount')
plt.ylabel('Profit')
plt.legend(title='Category')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```
# Replace 'Discount' and 'Category' with your actual column names
discount_col = 'Discount'
category_col = 'Category'
profit_col = 'Profit'

if discount_col not in df.columns or category_col not in df.columns or profit_col not in df.columns:
    print(f"Please replace '{discount_col}', '{category_col}', and '{profit_col}' with your actual column names.")
else:
    # Filter data where discount is greater than 0
    df_discounted = df[df[discount_col] > 0]

    # Group by category and sum profit for discounted items
    category_profit_discounted = df_discounted.groupby(category_col)[profit_col].sum().sort_values(ascending=False)

    # Create a bar graph of profit for discounted categories
    plt.figure(figsize=(10, 6))
    category_profit_discounted.plot(kind='bar', color='purple')
    plt.title('Total Profit by Category (Discounted Items Only)')
    plt.ylabel('Total Profit ($)')
    plt.xlabel('Category')
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()
```

8. List those product categories and sub categories which are given on discount and not making profit

```
# Replace column names if they are different
discount_col = 'Discount'
profit_col = 'Profit'
category_col = 'Category'
subcategory_col = 'Sub-Category'

if discount_col not in df.columns or profit_col not in df.columns or category_col not in df.columns or subcategory_col not in df.columns:
    print(f"Please replace column names if they are different: '{discount_col}', '{profit_col}', '{category_col}', '{subcategory_col}'.")
else:
    # Filter data for discounted items
    df_discounted = df[df[discount_col] > 0]

    # Filter discounted data for non-profitable items
    df_non_profitable_discounted = df_discounted[df_discounted[profit_col] <= 0]

    # Group by Category and Sub-Category and sum profit
    non_profitable_combinations_profit = df_non_profitable_discounted.groupby([category_col, subcategory_col])[profit_col].sum().sort_values()

    # Create a bar graph of profit for non-profitable discounted combinations
    plt.figure(figsize=(12, 6))
    non_profitable_combinations_profit.plot(kind='bar', color='red')
    plt.title('Total Profit for Non-Profitable Discounted Categories/Sub-Categories')
    plt.ylabel('Total Profit ($)')
    plt.xlabel('(Category, Sub-Category)')
    plt.xticks(rotation=90, ha='right')
    plt.tight_layout()
    plt.show()

    # Also print the list of non-profitable combinations for clarity
    non_profitable_combinations = df_non_profitable_discounted[[category_col, subcategory_col]].drop_duplicates()
    print("Product categories and sub-categories given on discount and not making profit:")
    for index, row in non_profitable_combinations.iterrows():
        print(f"Category: {row[category_col]}, Sub-Category: {row[subcategory_col]}")
```

```
# Replace column names if they are different
discount_col = 'Discount'
profit_col = 'Profit'
category_col = 'Category'
subcategory_col = 'Sub-Category'

if discount_col not in df.columns or profit_col not in df.columns or category_col not in df.columns or subcategory_col not in df.columns:
    print(f"Please replace column names if they are different: '{discount_col}', '{profit_col}', '{category_col}', '{subcategory_col}'.")
else:
    # Filter data for discounted items
    df_discounted = df[df[discount_col] > 0].copy() # Use .copy() to avoid SettingWithCopyWarning

    # Create a new column to identify non-profitable discounted items
    df_discounted['Profit Status'] = df_discounted[profit_col].apply(lambda x: 'Non-Profitable' if x <= 0 else 'Profitable')

    # Create a scatter plot
    plt.figure(figsize=(10, 6))
    sns.scatterplot(data=df_discounted, x=discount_col, y=profit_col, hue='Profit Status', alpha=0.7)
    plt.title('Profit vs. Discount for Discounted Items')
    plt.xlabel('Discount')
    plt.ylabel('Profit')
    plt.legend(title='Profit Status')
    plt.grid(True)
    plt.tight_layout()
    plt.show()

    # Also print the list of non-profitable combinations for clarity
    df_non_profitable_discounted = df_discounted[df_discounted[profit_col] <= 0]
    non_profitable_combinations = df_non_profitable_discounted[[category_col, subcategory_col]].drop_duplicates()
    print("Product categories and sub-categories given on discount and not making profit:")
    for index, row in non_profitable_combinations.iterrows():
        print(f"Category: {row[category_col]}, Sub-Category: {row[subcategory_col]}")
```
